### Visualizing The Gender Gap In College Degrees

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science',
             'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
                 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business',
              'Architecture']
```

```python
fig = plt.figure(figsize = (16,20))
dark_blue = (0/255,107/255,164/255)
orange = (255/255,128/255,14/255)

for i in range(0,16,3):
    ind = int(i/3)
#     print(ind)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[ind]], c = dark_blue, label = "Women", linewidth = 3)
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[ind]], c = orange, label = "Men", linewidth = 3)
    ax.set_title(stem_cats[ind])
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.tick_params(bottom = False, left = False, labelbottom = False)
    for v in ax.spines.values():
        v.set_visible(False)
    if ind == 0:
        ax.text(2004,80,"Women")
        ax.text(2004,15, "Men")
    if ind == 5:
        ax.text(2004,20,"Women")
        ax.text(2004,90, "Men")
        ax.tick_params(labelbottom = "on")
    ax.set_yticks([0,100])
    ax.axhline(50,c = (171/255,171/255,171/255) ,alpha = 0.3)


# plt.legend()
plt.show()
```
